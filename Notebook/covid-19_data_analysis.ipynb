## Importing the Modules

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import folium


print('Modules are Imported.')
```

## Importing COVID-19 Dataset

```python
df = pd.read_csv('..\Dataset\covid19_confirmed_dataset.csv')
df.head(10)
```

### Checking the shape of the Dataframe

```python
df.shape
```

## Deleting the useless columns

```python
df.drop(['Lat', 'Long'], axis=1, inplace=True)
df.head(10)
```

## Aggregating the rows by the country

```python
df_agg = df.groupby('Country/Region').sum()
df_agg.head(10)
```

```python
df_agg.shape
```

## Visualizing data related to different countries

```python
df_agg.loc['India'].plot()
df_agg.loc['US'].plot()
df_agg.loc['China'].plot()
plt.legend()
```

## Calculating the first derivative of the curves

```python
df_agg.loc['India'].diff().plot()
df_agg.loc['US'].diff().plot()
df_agg.loc['China'].diff().plot()
plt.legend()
```
